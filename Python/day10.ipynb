```python
def show_menu():
    print("\n -- to do list --")
    print('1 view tasks')
    print ('2 add task')
    print('3 remove task')
    print('4 exit')
```

```python
def load_tasks():
    try:
        with open("tasks.txt", "r") as f:
            return [line.strip() for line in f]
    except FileNotFoundError:
        return []
```

```python
def save_task(tasks):
    with open("tasks.txt","w") as f:
        for task in tasks:
            f.write(task +'\n')
```

```python
tasks = load_tasks()

on = True

while on:
    show_menu()
    choice = input('enter your choice: ')

    if choice == "1":
        if not tasks:
            print("No tasks to display.")
        else:
            for i, task in enumerate(tasks,1):
                print(f"{i}.{task}")
    elif choice == "2":
        task = input("Enter the task: ")
        tasks.append(task)
        save_task(tasks)
        print('Task added successfully.')
    elif choice =="3":
        for i, task in enumerate(tasks,1):
            print(f"{i}. {task}")
        num = int(input("enter the number of the task you want to delete: "))
        if 1 <= num <= len(tasks):
            removed = tasks.pop(num-1)
            save_task(task)
            print(f' the task {removed} has been deleted.')
        else:
            print('Invalid choice, please try again.')
            
    elif choice == "4":
        print("Goodbye!")
        on = False
    else:
        print("invalid input")
```
