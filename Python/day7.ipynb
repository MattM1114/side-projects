```python
# password generator python
import random as rand 
import string as st

length = int(input("enter the length of the pass word you want: "))

chars = st.ascii_letters + st.digits + st.punctuation
passw = "".join(rand.choice(chars) for _ in range(length))
print("this is your password: ",passw)

with open("passwords.txt","a")as f:
    f.write(passw+"\n")
```

```python
import time 
sec = int(input("Enter your time in seconds: "))

while sec >0:
    print(f"time left:{sec} seconds")
    time.sleep(1)
    sec -=1
    
print('Time up!')
```
