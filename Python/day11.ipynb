```python
import json
import os

File = "Tasks.json"

# this will load the tasks from json file
def load_tasks():
    if not os.path.exists(File):
        return []
    
    with open(File,"r") as f:
        return json.load(f)

# this will save the tasks
def save_tasks(tasks):
    with open(File,"w") as f:
        json.dump(tasks,f,indent=4)
```

```python
# this will add tasks
def add_task(tasks):
    title = input("Enter the task titile: ").strip()
    
    if not title:
        print("task can't be empty.")
        return
    
    task = {
        "id": len(tasks)+1,
        "title": title,
        "completed": False

    }
    tasks.append(task)
    print("task added.")
```

```python
# this will complete tasks
def complete_task(tasks):
    try:
        task_1d = int(input("Enter the task ID to complete: "))
    except ValueError:
        print("Invalid number.")
        return
    
    for task in tasks:
        if task["id"] == task_1d:
            task["completed"] = True
            print(f"task {task['title']} completed")
            return
        
    print("Task not found.")
```

```python
# this will display all the tasks
def display(tasks):
    if not tasks:
        print("No tasks.")
        return

    for task in tasks:
        status= "complete" if task["completed"] else "incomplete"

        print(f"{task['id']}. {task['title']} [{status}]")
```

```python
""" this is the main function that will run the progarm
we need to display all the functions then we use if statement to check the user's choice
and we use a loop to keep the program running until the they press 4"""

def main():
    tasks = load_tasks()
    
    while True:
        print("\n1. view Tasks \n2 Add task \n3 Complete Task\n4 Exit")
        
        choice = input("Enter your choice: ")
        
        if choice == "1":
            display(tasks)
        
        elif choice == "2":
            add_task(tasks)
        
        elif choice == "3":
            complete_task(tasks)
            save_tasks(tasks)
        elif choice == "4":
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Please try again")

if __name__ == "__main__":
    main()
```
