```python
#the 1st function is celsius to fahrenheit
def cel_to_fah(float:temp):
    fahrenheit = (temp * 9/5)+32
    return fahrenheit
```

```python
#the 2nd function is fahrenheit to celsius
def fah_to_cel(float:temp):
    celsius = (temp - 32) * 5/9
    return celsius
```

```python
#the 3rd function is celsius to kelvin
def cel_to_kel(float:temp):
    kelvin = temp + 273.15
    return kelvin
```

```python
#the 4th function is kelvin to celsius
def kel_to_cel(float:temp):
    celsius = temp - 273.15
    return celsius
```

```python
#the 5th function is fahrenheit to kelvin
def fah_to_kel(float:temp):
    kelvin = (temp -32)*9/5 + 273.15
    return kelvin
```

```python
#the 6th function is kelvin to fahrenheit
def kel_to_fah(float:temp):
    fahrenheit = (temp - 273.15)9/5 + 32
    return fahrenheit
```

```python
#Main program to take user input and call the appropriate function
temp = float(input("Enter the temperature value: "))
input_unit = input("Enter the input unit (C, F, K): ").upper()
output_unit = input("Enter the output unit (C, F, K): ").upper()


if input_unit == "C"and output_unit == "F":
    result = cel_to_fah(temp)
elif input_unit == "F" and output_unit == "C":
    result = fah_to_cel(temp)
elif input_unit == 'C'and output_unit == "K":
    result = cel_to_kel(temp)
elif input_unit == "K"and output_unit == 'C':
    result = kel_to_cel(temp)
elif input_unit == 'F'and output_unit == 'K':
    result = fah_to_kel(temp)
elif input_unit == 'K'and output_unit == 'F':
    result = kel_to_fah(temp)
else:
    result = "Invalid input or output unit."


if isinstance(result, float):
    print(f"The converted temperature is: {result} {output_unit}")
else:
    print(result)
```
