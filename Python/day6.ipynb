```python
import random

secret = random.randint(1,100)
guess = None
attempts = 0

print("Guess the number between 1 and 100!")

while guess != secret:
    guess = int(input("Enter your guess: "))
    attempts += 1
    
    if guess < secret:
        print("Too low!")
    elif guess > secret:
        print('Too high!')
    else:
        print(f"Correct! you guessed it in {attempts} attempts.")
```
