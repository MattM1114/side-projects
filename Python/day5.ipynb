```python
# 1st project is odd or even today


even = 0
odd = 0 


for i in range(10):
    num = int(input("enter a integer: "))
    if num % 2 == 0:
        print(f"{num} is even")
        even +=1
    else:
        print(f"{num} is odd")
        odd +=1

print(f"total even numbers: {even} and total odd numbers: {odd} out of 10 numbers")
```

```python
# challenge we are checking if a number is prime or not

num3 = int(input("enter a integer: "))

def  is_prime(n):
    if n > 1:
        for i in range(2, int(n)):
            if (n % i) == 0:
                return False
        else:
            return True

print(f"Prime: {is_prime(num3)}")
```
