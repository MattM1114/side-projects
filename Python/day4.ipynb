```python
# to set up a madlib generator we need to  import random library
import random
```

```python
# the inputs required for madlib generator are the following
name = input("Enter your name: ")
food = input("Enter a food item: ")
place = input("Enter a place: ")
animal = input("Enter an animal: ")
adjective = input("Enter an adjective: ")
verb = input("Enter a verb: ")
```

```python
#we need a list of madlib templates to choose from

stories =[f"{name} went to {place}and fought a {animal} with a {food}."
        ,f'In {place}, a {adjective} {animal} started to {verb} near {name}.',
        f'{name} cooked a delicious {food} for a {animal} and everyone was {adjective}.',
        f'{name} loves to {verb} while eating {food} in {place}.',
        f'One day, {name} saw a {adjective} {animal} in {place} and decided to {verb} with it.'
        ]

```

```python
print("Here is your madlib story:")
story = random.choice(stories)
print(story)
with open("story.txt","w") as f:
    f.write(story)
```
