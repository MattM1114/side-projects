```python
def add(a,b):
    return a+b
#this is the add function
```

```python
def power(a,b):
    return a**b
#this is the power function
```

```python
def subtract(a,b):
    return a-b
#this is the subtract function
```

```python
def multiply(a,b):
    return a*b
# this is the multiply function
```

```python
def divide(a, b):
    if b == 0:
        return "Error: Division by zero is not allowed."
    else:
        return a / b
# this is the divide function
```

```python
print("simple python calculator.")
print("Select operation.")
print("1.Add")
print("2.Subtract")
print("3.Multiply")
print("4.Divide")
print("5.Power")
choice = input("Enter choice(1/2/3/4/5): ")
#this is the user input for the operation
```

```python
num1 = float(input("Enter the 1st number:"))
num2 = float(input("Enter the 2nd number:"))
#this is the user input for the numbers to be calculated
    
```

```python

if choice == '1':
    print(add(num1,num2))
elif choice == '2':
    print(subtract(num1,num2))
elif choice =='3':
    print(multiply(num1,num2))
elif choice =='4':
    print(divide(num1,num2))
elif choice == '5':
    print(power(num1,num2))
else:
    print("Invalid input")
#this is the operation execution based on user input
```

```python
while True:
    again = input("Do you want to perform another calculation? (yes/no): ")
    if again.lower() in ('yes', 'y'):
        num1 = float(input("Enter the 1st number:"))
        num2 = float(input("Enter the 2nd number:"))
        choice = input("Enter choice(1/2/3/4/5): ")
        if choice == '1':
            print(add(num1,num2))
        elif choice == '2':
            print(subtract(num1,num2))
        elif choice =='3':
            print(multiply(num1,num2))
        elif choice =='4':
            print(divide(num1,num2))
        elif choice == '5':
            print(power(num1,num2))
        else:
            print("Invalid input")
    else:
        print("Thank you for using the calculator. Goodbye!")
        break
    #this loop allows the user to perform multiple calculations until they choose to exit
```
