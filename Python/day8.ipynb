```python
import random


choices = ["rock","paper","scissors"]

score = 0
while True:
    

    player = int(input("please pick between 0 rock,1 paper, and 2 scissors: "))

    computer = random.randint(0,2)

    if  choices[player] == choices[computer]:
        print(f"you both chose {choices[player]} and it was a tie ")
        
    elif ((choices[player] == "rock" and choices[computer] == "scissors") 
        or (choices[player] == "paper" and choices[computer] == "rock") 
        or (choices[player] == "scissors" and choices[computer] =="paper")):
        print (f"you chose {choices[player]} and the the computer chose {choices[computer]} you win ")
        score += 1
    else:
        print(f"you chose {choices[player]} and the computer chose {choices[computer]} you lose")

    again = input("Do you want to play again? (y/n) ")
    print(f"your score is {score}")
    if again.lower() != "y":
        break
```

```python
from collections import Counter

text = input("please enter a sentence: ")

words = text.split()

w_cont = len(words)

word_f = Counter(words)
most_c = word_f.most_common(1)

print(f"Total word count is {w_cont}")
print(f"Most common word is {most_c}")
```
